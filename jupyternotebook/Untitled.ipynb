```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Desktop/碩論/數據資料/定位位移/3DReprojection_differentmethod.csv')
```

```python
n=5
index = np.arange(5)
width = 0.2  
```

```python
AutoReg = df[df.Method == "AutoReg"].values.tolist()
Kalman = df[df.Method == "Kalman"].values.tolist()
Baseline = df[df.Method == "Baseline"].values.tolist()
Ours = df[df.Method == "Ours"].values.tolist()
Baseline=Baseline[0][1:]
AutoReg=AutoReg[0][1:]
Kalman=Kalman[0][1:]
Ours=Ours[0][1:]
```

```python
x1=[40, 60, 80, 100, 200]          #X 軸 (第一組)
x2=[p + width for p in x1]                 #X 軸 (第二組)
x3=[p + 2*width for p in x1]                 #X 軸 (第三組)
x4=[p + 3*width for p in x1]                 #X 軸 (第四組)
x5=[p + 4*width for p in x1]                 #X 軸 (第四組)
```

```python
plt.figure(dpi = 200)
plt.bar(index+0.75, Ours, label='LARR', width=0.25)  #繪製長條圖
plt.bar(index+0.5, Kalman, label='Kalman', width=0.25)  #繪製長條圖
plt.bar(index+0.25, AutoReg, label='AutoReg', width=0.25)  #繪製長條圖
plt.bar(index, Baseline, label='No Prediction', width=0.25)  #繪製長條圖
plt.xticks(index+ .5/ 2 ,("40", "60", "80", "100", "200"))  #設定 X 軸刻度標籤

plt.legend()                               #顯示圖例
plt.xlabel('Motion to Photon latency (ms)')   #設定 X 軸標籤
plt.ylabel('Position offset')          #設定 Y 軸標籤
plt.yscale('log')
plt.savefig('./Desktop/碩論/圖/Positionerror_compare_withours.png',pad_inches = 0.0)
plt.show()
```
