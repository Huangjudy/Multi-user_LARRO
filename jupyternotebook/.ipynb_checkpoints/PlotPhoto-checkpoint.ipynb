```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./Desktop/碩論/圖/3DReprojection.csv')
```

```python
print (df)
```

```python
df_walk_reprojection_error_2D = df['reprojection_error_2D']
df_walk_reprojection_error_2D.dropna(axis=0, how='any', inplace=True)
print(df_walk_reprojection_error_2D)
```

```python
df_walk_reprojection_error_3D = df['reprojection_error_3D']
df_walk_reprojection_error_3D.dropna(axis=0, how='any', inplace=True)
print(df_walk_reprojection_error_3D)
```

```python
df_goandback_reprojection_error_2D = df['reprojection_error_2D']
df_goandback_reprojection_error_2D.dropna(axis=0, how='any', inplace=True)
print(df_goandback_reprojection_error_2D)
```

```python
df_goandback_reprojection_error_3D = df['reprojection_error_3D']
df_goandback_reprojection_error_3D.dropna(axis=0, how='any', inplace=True)
print(df_goandback_reprojection_error_3D)
```

```python
import matplotlib.font_manager as fm
f1 = fm.FontProperties(size=10)
plt.figure( figsize = (9,6) , dpi = 200)
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.2)

#plt.subplot(121)
plt.plot( df_walk_reprojection_error_2D , color='darkorange', linestyle='-', linewidth=2,label=r'2D (normalize)')
plt.plot( df_walk_reprojection_error_3D , color='darkgreen', linestyle='-', linewidth=2, label=r'3D (m)')
plt.ylabel('offset', fontsize=16)
plt.xlabel('Frame', fontsize=16)
plt.xlim(0, len(df_walk_reprojection_error_2D))
plt.tick_params(axis='both', labelsize=16)
plt.ylim(0,)
plt.title(r'Positioning offset in different dimensions', fontsize=20)
plt.legend(prop={'size': 16},fontsize="x-large")
plt.grid(True)


# plt.subplot(122)
# plt.plot( df_goandback_reprojection_error_2D , color='darkorange', linestyle='-', linewidth=3,label=r'Position_error_2D (normalize)')
# plt.plot( df_goandback_reprojection_error_3D , color='darkgreen', linestyle='-', linewidth=3,label=r'Position_error_3D (m)')
# plt.ylabel('Position offset', fontsize=16)
# plt.xlabel('Frame', fontsize=16)
# plt.xlim(0, len(df_goandback_reprojection_error_2D))
# plt.tick_params(axis='both', labelsize=16)
# plt.ylim(0,1.5)
# plt.title(r'Walk slowly (close to the object)', fontsize=20)
# plt.legend(prop={'size': 16},fontsize="x-large")
# plt.grid(True)

plt.savefig('./Desktop/碩論/圖/Positionoffset_detail.png',pad_inches = 0.0)
```
