```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/IPC_LAB/Desktop/data/multi_data.csv")
```

```python
n=5
index = np.arange(5)
width = 0.2
```

```python
baseline = df[df.method == "baseline"].values.tolist()
no_opt = df[df.method == "no_opt"].values.tolist()
opt = df[df.method == "opt"].values.tolist()
#Ours = df[df.method == "Ours"].values.tolist()
baseline=baseline[0][1:]
no_opt=no_opt[0][1:]
opt=opt[0][1:]
#Ours=Ours[0][1:]
```

```python
x1=[1, 2, 3, 4, 5]          #X 軸 (第一組)
x2=[p + width for p in x1]                 #X 軸 (第二組)
x3=[p + 2*width for p in x1]                 #X 軸 (第三組)
x4=[p + 3*width for p in x1]                 #X 軸 (第四組)
x5=[p + 4*width for p in x1]                 #X 軸 (第四組)
```

```python
plt.figure(dpi = 200)
#plt.bar(index+0.75, Ours, label='LARR', width=0.25)  #繪製長條圖
plt.bar(index+0, baseline, label='Baseline', width=0.25)  #繪製長條圖
plt.bar(index+0.25, no_opt, label='Multi-user LARR', width=0.25)  #繪製長條圖
plt.bar(index+0.5, opt, label='Multi-user LARRO', width=0.25)  #繪製長條圖
#plt.xticks(index+ 0.25 ,ticks=("1", "2", "3", "4", "5") ) #設定 X 軸刻度標籤
plt.xticks(index+ 0.25,x1)
plt.legend()                               #顯示圖例
plt.xlabel('Number of users')   #設定 X 軸標籤
plt.ylabel('NetworkUsage (Kb)')          #設定 Y 軸標籤
#plt.yscale('log')
plt.savefig('E:/picture/Multi-LARR.jpg',pad_inches = 0.0)
plt.show()
```
