```python
df = df[~df['Protocol'].str.contains('STUN')]
df = df[~df['Protocol'].str.contains('TCP')]
df_data = df[~df['Protocol'].str.contains('HTTP')]
df_protocol = df_data.groupby('Protocol').Source.count()
df_protocol.plot(kind='bar')
```

```python
df_ClienttoServer = df_data[df_data['Source'] == '140.115.152.169']
#df_ClienttoServer = df_data[df_data['Source'] == '140.115.197.252']
df_ServertoClient = df_data[df_data['Source'] == '140.115.152.173']
df_ServertoClient
```

```python
#時間界線
lpper = 5
upper = 240
```

```python
df_ClienttoServer = df_ClienttoServer[ df_ClienttoServer['Time'].between(lpper,upper)]
df_ClienttoServer
```

```python
df_ServertoClient = df_ServertoClient[ df_ServertoClient['Time'].between(lpper,upper) ]
df_ServertoClient
```

```python
list_ServertoClientnetworkusage = []
list_axisxServertoClient = []
for i in range(lpper,upper):
    df_part = df_ServertoClient[ df_ServertoClient['Time'].between(i,i+1) ]
    totally_byte_per_second = df_part['Length'].sum()  #每秒多少bytes
    totally_bits_per_second = totally_byte_per_second * 8 #每秒多少bits
    Kbits_per_second = totally_bits_per_second / 1024
    list_ServertoClientnetworkusage.append(Kbits_per_second)
    list_axisxServertoClient.append(i)
    print( '使用了 {} Kb' .format(Kbits_per_second))
```

```python
list_ClienttoServernetworkusage = []
list_axisxClienttoServer = []
for i in range(lpper,upper):
    df_part = df_ClienttoServer[ df_ClienttoServer['Time'].between(i,i+1) ]
    totally_byte_per_second = df_part['Length'].sum()  #每秒多少bytes
    totally_bits_per_second = totally_byte_per_second * 8 #每秒多少bits
    Kbits_per_second = totally_bits_per_second / 1024
    list_ClienttoServernetworkusage.append(Kbits_per_second)
    list_axisxClienttoServer.append(i)
    print( '使用了 {} Kb' .format(Kbits_per_second))
```

```python
fig = plt.figure()
#list_ServertoClientnetworkusage[8:15] = list_ServertoClientnetworkusage[0:7]
plt.plot(list_axisxClienttoServer, list_ClienttoServernetworkusage, 'r')
plt.plot(list_axisxServertoClient, list_ServertoClientnetworkusage, 'g')
plt.legend(('Client to Server', 'Server to Client'), loc='best') # 畫圖例及決定位置
plt.title('Multiplane Streaming')
plt.xlabel('Sample number (sec)')
plt.ylabel('NetworkUsage (Kb)')
plt.show()
```

```python
mean_Client_to_Server = sum(list_ClienttoServernetworkusage) / len(list_axisxClienttoServer)
mean_Server_to_Client = sum(list_ServertoClientnetworkusage) / len(list_axisxServertoClient)
print('平均每秒串流至Client的影像資料為 {} Kb'.format(mean_Server_to_Client))
print('平均每秒傳輸至Server的控制資料為 {} Kb'.format(mean_Client_to_Server))
```
